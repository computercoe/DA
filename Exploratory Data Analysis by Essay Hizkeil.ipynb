```python
numerical_columns = ['Height_(cm)', 'Weight_(kg)', 'BMI']

for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[column], kde=True)
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
categorical_columns = ['Alcohol_Consumption', 'Fruit_Consumption', 'Green_Vegetables_Consumption', 'FriedPotato_Consumption']

for column in categorical_columns:
    plt.figure(figsize=(10, 8))
    sns.countplot(data[column])
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
inary_columns = ['Exercise', 'Heart_Disease', 'Skin_Cancer',
                  'Other_Cancer', 'Depression', 'Diabetes', 'Arthritis',
                  'Smoking_History']

for column in binary_columns:
    data[column] = data[column].map({'Yes': 1, 'No': 0})

# Calculate the percentage of 'Yes' for each binary column
percentage_yes = data[binary_columns].mean() * 100

# Plot the percentage of 'Yes' for each binary column
plt.figure(figsize=(12, 8))
sns.barplot(x=percentage_yes.index, y=percentage_yes.values)
plt.xlabel('Binary Columns')
plt.ylabel('Percentage of Yes')
plt.title('Percentage of Yes for Binary Columns')
plt.xticks(rotation=45)
plt.show()
```

```python
general_health_categories = ['Poor', 'Good', 'Fair', 'Excellent', 'Very Good']

# Count and plot the frequency of each category in General_Health column
category_counts = data['General_Health'].value_counts()
plt.figure(figsize=(12, 8))
sns.barplot(x=category_counts.index, y=category_counts.values, order=general_health_categories)
plt.xlabel('General_Health')
plt.ylabel('Frequency')
plt.title('Frequency of Categories in General_Health')
plt.xticks(rotation=45)
plt.show()

```

```python
categorical_columns = ['General_Health', 'Exercise', 'Heart_Disease', 'Skin_Cancer', 'Other_Cancer',
                       'Depression', 'Diabetes', 'Arthritis', 'Sex', 'Age_Category']
category_mapping = {
    'General_Health': ['Poor', 'Fair', 'Good', 'Very Good', 'Excellent'],
    'Exercise': ['Yes', 'No'],
    'Heart_Disease': ['Yes', 'No'],
    'Skin_Cancer': ['Yes', 'No'],
    'Other_Cancer': ['Yes', 'No'],
    'Depression': ['Yes', 'No'],
    'Diabetes': ['Yes', 'No'],
    'Arthritis': ['Yes', 'No'],
    'Sex': ['Female', 'Male'],
    'Age_Category': ['<18', '18-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-65','65-69', '70-74', '75-79', '80+']
}

# Perform EDA for each categorical variable
for column in categorical_columns:
    # Count the frequency of each category
    category_counts = data[column].value_counts()

    # Sort the categories based on the defined mapping
    sorted_categories = sorted(category_mapping[column], key=lambda x: category_counts.get(x, 0), reverse=True)

    # Plot the bar chart
    plt.figure(figsize=(10, 8))
    sns.countplot(data=data, x=column, order=sorted_categories)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Frequency of Categories in {column}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
numerical_columns = ['Height_(cm)', 'Weight_(kg)', 'BMI', 'Alcohol_Consumption', 'Fruit_Consumption', 'Green_Vegetables_Consumption', 'FriedPotato_Consumption']

for column in numerical_columns:
    # Plot the histogram
    plt.figure(figsize=(10, 8))
    sns.histplot(data=data, x=column)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
numerical_columns = ['Height_(cm)', 'Weight_(kg)', 'BMI', 'Alcohol_Consumption', 'Fruit_Consumption', 'Green_Vegetables_Consumption', 'FriedPotato_Consumption']
correlation_matrix = data[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap for Numerical Variables')
plt.show()
```

```python
for column in numerical_columns:
    plt.figure(figsize=(10, 8))
    sns.boxplot(data=data, x='Age_Category', y=column)
    plt.xlabel('Age_Category')
    plt.ylabel(column)
    plt.title(f'Box Plot of {column} by Age_Category')
    plt.xticks(rotation=45)
    plt.show()
```

```python
categorical_columns = ['General_Health']
category_mapping = { 'General_Health': ['Poor', 'Fair', 'Good', 'Very Good', 'Excellent']
}

# Plot the average alcohol consumption for each category
for column in categorical_columns:
    plt.figure(figsize=(10, 8))
    sns.barplot(data=data, x=column, y='Alcohol_Consumption', order=category_mapping[column])
    plt.xlabel(column)
    plt.ylabel('Average Alcohol Consumption')
    plt.title(f'Average Alcohol Consumption by {column}')
    plt.show()
```

```python
categorical_column = 'General_Health'
category_order = ['Poor', 'Fair', 'Good', 'Very Good', 'Excellent']

# Define the numerical columns of interest
numerical_columns = ['Fruit_Consumption', 'Green_Vegetables_Consumption', 'FriedPotato_Consumption']

# Plot the distribution of numerical variables for each category of general health
for column in numerical_columns:
    plt.figure(figsize=(10, 8))
    sns.boxplot(data=data, x=categorical_column, y=column, order=category_order)
    plt.xlabel(categorical_column)
    plt.ylabel(column)
    plt.title(f'Distribution of {column} by {categorical_column}')
    plt.xticks(rotation=45)
    plt.show()
```
